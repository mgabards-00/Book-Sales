# DATA SUMMARY

## Insigths
- The data shows that **Science Fiction** has the most garnered total revenue with a percentage of **42.5%** followed by **Non-Fiction** with **31.5%**
- The **Fiction** genre has the least total revenue with **26.0%**
- The revenue percentage for every genre shows consumers has preferences depending on each trend and market value.

## Conclusion

1. Science Fiction has the most revenue depicting from the chart shown. Meaning we should boost the genre and provide prior promotion to earn more revenue as this is the choice of most consumers.

2. Fiction on the otherhand might require a boost on sales and promotions to increase its revenue.

3. Observing the trend movements per genre will help boost revenue from time to time and will give us insights and strategies to play along the market with the consumers.
